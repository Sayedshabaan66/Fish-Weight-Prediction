**Fish Weight Prediction**

```python
import pandas as pd
```

```python
fish= pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Fish.csv')
```

```python
fish.head()
```

```python
fish.describe()
```

```python
fish.info()
```

```python
fish.columns
```

```python
y = fish['Weight']
```

```python
x = fish[['Category', 'Height', 'Width', 'Length1',
       'Length2', 'Length3']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, random_state= 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Assume 'data' is your dataset with relevant features and predicted fish weights
# You can replace it with your actual dataset
# For simplicity, let's assume you have features 'length' and 'predicted_weight'
data = {'length': [15, 20, 25, 30, 35],
        'predicted_weight': [0.5, 1.2, 2.0, 3.5, 4.8]}

# Create a DataFrame (you can replace this with your actual DataFrame)
import pandas as pd
df = pd.DataFrame(data)

# Scatter plot for length and predicted weight
plt.figure(figsize=(8, 6))
sns.scatterplot(x='length', y='predicted_weight', data=df)
plt.title('Fish Weight Prediction Visualization')
plt.xlabel('Fish Length (cm)')
plt.ylabel('Predicted Weight (kg)')
plt.show()
```

```python
y_predict= model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
r2_score(y_test, y_predict)
```
